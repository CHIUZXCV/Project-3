```python
d = []
train_years = [2015, 2016, 2017, 2018]

for i in train_years:
    r = requests.get(f'https://v3.football.api-sports.io/fixtures?season={i}&league=39', headers=headers)

    for i in range(50):
        if r.json()['response'][i]['teams']['home']['winner'] == True:
            d.append(
            {
                'date': r.json()['response'][i]['fixture']['date'],
                'fixture_id': r.json()['response'][i]['fixture']['id'],
                'home_team': r.json()['response'][i]['teams']['home']['name'],
                'away_team': r.json()['response'][i]['teams']['away']['name'],
                'home_goals': r.json()['response'][i]['goals']['home'],
                'away_goals': r.json()['response'][i]['goals']['away'],
                'result': 1
            }
        )
        elif r.json()['response'][i]['teams']['home']['winner'] == False:
            d.append(
            {
                'date': r.json()['response'][i]['fixture']['date'],
                'fixture_id': r.json()['response'][i]['fixture']['id'],
                'home_team': r.json()['response'][i]['teams']['home']['name'],
                'away_team': r.json()['response'][i]['teams']['away']['name'],
                'home_goals': r.json()['response'][i]['goals']['home'],
                'away_goals': r.json()['response'][i]['goals']['away'],
                'result': -1
            }
        )
        else:
            d.append(
            {
                'date': r.json()['response'][i]['fixture']['date'],
                'fixture_id': r.json()['response'][i]['fixture']['id'],
                'home_team': r.json()['response'][i]['teams']['home']['name'],
                'away_team': r.json()['response'][i]['teams']['away']['name'],
                'home_goals': r.json()['response'][i]['goals']['home'],
                'away_goals': r.json()['response'][i]['goals']['away'],
                'result': 0
            }
        )
data_2015 = pd.DataFrame(d)
data_2015.set_index('date', inplace=True)
```

```python
# Initialize empty lists to be used in a for-loop that will fill the home team's data

home_shots_on_goal = []
home_shots_off_goal = []
home_total_shots = []
home_blocked_shots = []
home_shots_insidebox = []
home_shots_outsidebox = []
home_fouls = []
home_cornerkicks = []
home_offsides = []
home_possession = []
home_yellowcards = []
home_redcards = []
home_gk_saves = []
home_pass_percentage = []

# Initialize empty lists to be used in a for-loop that will fill the away team's data

away_shots_on_goal = []
away_shots_off_goal = []
away_total_shots = []
away_blocked_shots = []
away_shots_insidebox = []
away_shots_outsidebox = []
away_fouls = []
away_cornerkicks = []
away_offsides = []
away_possession = []
away_yellowcards = []
away_redcards = []
away_gk_saves = []
away_pass_percentage = []

# for-loop that request statistics for each fixture in a season and stores it in
# the corresponding list

for i in data_2015['fixture_id']:
    stats = requests.get(f"https://v3.football.api-sports.io/fixtures/statistics?fixture={i}", headers=headers)
    if stats.json()['response'] != []:
        if stats.json()['response'][0]['statistics'][0]['value'] == None:
            home_shots_on_goal.append(0)
        else:
            home_shots_on_goal.append(stats.json()['response'][0]['statistics'][0]['value'])
        if stats.json()['response'][0]['statistics'][1]['value'] == None:
            home_shots_off_goal.append(0)
        else:
            home_shots_off_goal.append(stats.json()['response'][0]['statistics'][1]['value'])
        if stats.json()['response'][0]['statistics'][2]['value'] == None:
            home_total_shots.append(0)
        else:
            home_total_shots.append(stats.json()['response'][0]['statistics'][2]['value'])
        if stats.json()['response'][0]['statistics'][3]['value'] == None:
            home_blocked_shots.append(0)
        else:
            home_blocked_shots.append(stats.json()['response'][0]['statistics'][3]['value'])
        if stats.json()['response'][0]['statistics'][4]['value'] == None:
            home_shots_insidebox.append(0)
        else:
            home_shots_insidebox.append(stats.json()['response'][0]['statistics'][4]['value'])
        if stats.json()['response'][0]['statistics'][5]['value'] == None:
            home_shots_outsidebox.append(0)
        else:
            home_shots_outsidebox.append(stats.json()['response'][0]['statistics'][5]['value'])
        if stats.json()['response'][0]['statistics'][6]['value'] == None:
            home_fouls.append(0)
        else:
            home_fouls.append(stats.json()['response'][0]['statistics'][6]['value'])
        if stats.json()['response'][0]['statistics'][7]['value'] == None:
            home_cornerkicks.append(0)
        else:
            home_cornerkicks.append(stats.json()['response'][0]['statistics'][7]['value'])
        if stats.json()['response'][0]['statistics'][8]['value'] == None:
            home_offsides.append(0)
        else:
            home_offsides.append(stats.json()['response'][0]['statistics'][8]['value'])
        home_possession.append(float(stats.json()['response'][0]['statistics'][9]['value'].strip('%')) * 0.01)
        if stats.json()['response'][0]['statistics'][10]['value'] == None:
            home_yellowcards.append(0)
        else:
            home_yellowcards.append(stats.json()['response'][0]['statistics'][10]['value'])
        if stats.json()['response'][0]['statistics'][11]['value'] == None:
            home_redcards.append(0)
        else:
            home_redcards.append(stats.json()['response'][0]['statistics'][11]['value'])
        if stats.json()['response'][0]['statistics'][12]['value'] == None:
            home_gk_saves.append(0)
        else:
            home_gk_saves.append(stats.json()['response'][0]['statistics'][12]['value'])
        home_pass_percentage.append(stats.json()['response'][0]['statistics'][14]['value'] / stats.json()['response'][0]['statistics'][13]['value'])

# Continue loop for away team
    
        if stats.json()['response'][1]['statistics'][0]['value'] == None:
            away_shots_on_goal.append(0)
        else:
            away_shots_on_goal.append(stats.json()['response'][1]['statistics'][0]['value'])
        if stats.json()['response'][1]['statistics'][1]['value'] == None:
            away_shots_off_goal.append(0)
        else:
            away_shots_off_goal.append(stats.json()['response'][1]['statistics'][1]['value'])
        if stats.json()['response'][1]['statistics'][2]['value'] == None:
            away_total_shots.append(0)
        else:
            away_total_shots.append(stats.json()['response'][1]['statistics'][2]['value'])
        if stats.json()['response'][1]['statistics'][3]['value'] == None:
            away_blocked_shots.append(0)
        else:
            away_blocked_shots.append(stats.json()['response'][1]['statistics'][3]['value'])
        if stats.json()['response'][1]['statistics'][4]['value'] == None:
            away_shots_insidebox.append(0)
        else:
            away_shots_insidebox.append(stats.json()['response'][1]['statistics'][4]['value'])
        if stats.json()['response'][1]['statistics'][5]['value'] == None:
            away_shots_outsidebox.append(0)
        else:
            away_shots_outsidebox.append(stats.json()['response'][1]['statistics'][5]['value'])
        if stats.json()['response'][1]['statistics'][6]['value'] == None:
            away_fouls.append(0)
        else:
            away_fouls.append(stats.json()['response'][1]['statistics'][6]['value'])
        if stats.json()['response'][1]['statistics'][7]['value'] == None:
            away_cornerkicks.append(0)
        else:
            away_cornerkicks.append(stats.json()['response'][1]['statistics'][7]['value'])
        if stats.json()['response'][1]['statistics'][8]['value'] == None:
            away_offsides.append(0)
        else:
            away_offsides.append(stats.json()['response'][1]['statistics'][8]['value'])
        away_possession.append(float(stats.json()['response'][1]['statistics'][9]['value'].strip('%')) * 0.01)
        if stats.json()['response'][1]['statistics'][10]['value'] == None:
            away_yellowcards.append(0)
        else:
            away_yellowcards.append(stats.json()['response'][1]['statistics'][10]['value'])
        if stats.json()['response'][1]['statistics'][11]['value'] == None:
            away_redcards.append(0)
        else:
            away_redcards.append(stats.json()['response'][1]['statistics'][11]['value'])
        if stats.json()['response'][1]['statistics'][12]['value'] == None:
            away_gk_saves.append(0)
        else:
            away_gk_saves.append(stats.json()['response'][1]['statistics'][12]['value'])
        away_pass_percentage.append(stats.json()['response'][1]['statistics'][14]['value'] / stats.json()['response'][1]['statistics'][13]['value'])
    else:
        home_shots_on_goal.append(np.nan)
        home_shots_off_goal.append(np.nan)
        home_total_shots.append(np.nan)
        home_blocked_shots.append(np.nan)
        home_shots_insidebox.append(np.nan)
        home_shots_outsidebox.append(np.nan)
        home_fouls.append(np.nan)
        home_cornerkicks.append(np.nan)
        home_offsides.append(np.nan)
        home_possession.append(np.nan)
        home_yellowcards.append(np.nan)
        home_redcards.append(np.nan)
        home_gk_saves.append(np.nan)
        home_pass_percentage.append(np.nan)
        away_shots_on_goal.append(np.nan)
        away_shots_off_goal.append(np.nan)
        away_total_shots.append(np.nan)
        away_blocked_shots.append(np.nan)
        away_shots_insidebox.append(np.nan)
        away_shots_outsidebox.append(np.nan)
        away_fouls.append(np.nan)
        away_cornerkicks.append(np.nan)
        away_offsides.append(np.nan)
        away_possession.append(np.nan)
        away_yellowcards.append(np.nan)
        away_redcards.append(np.nan)
        away_gk_saves.append(np.nan)
        away_pass_percentage.append(np.nan)
    

# Add columnds of home team data to the season's dataframe

data_2015['home_shots_on_goal'] = home_shots_on_goal
data_2015['home_shots_off_goal'] = home_shots_off_goal
data_2015['home_total_shots'] = home_total_shots
data_2015['home_blocked_shots'] = home_blocked_shots
data_2015['home_shots_insidebox'] = home_shots_insidebox
data_2015['home_shots_outsidebox'] = home_shots_outsidebox
data_2015['home_fouls'] = home_fouls
data_2015['home_cornerkicks'] = home_cornerkicks
data_2015['home_offsides'] = home_offsides
data_2015['home_possession'] = home_possession
data_2015['home_yellowcards'] = home_yellowcards
data_2015['home_redcards'] = home_redcards
data_2015['home_gk_saves'] = home_gk_saves
data_2015['home_pass_percentage'] = home_pass_percentage

# Add columnds of away team data to the season's data frame

data_2015['away_shots_on_goal'] = away_shots_on_goal
data_2015['away_shots_off_goal'] = away_shots_off_goal
data_2015['away_total_shots'] = away_total_shots
data_2015['away_blocked_shots'] = away_blocked_shots
data_2015['away_shots_insidebox'] = away_shots_insidebox
data_2015['away_shots_outsidebox'] = away_shots_outsidebox
data_2015['away_fouls'] = away_fouls
data_2015['away_cornerkicks'] = away_cornerkicks
data_2015['away_offsides'] = away_offsides
data_2015['away_possession'] = away_possession
data_2015['away_yellowcards'] = away_yellowcards
data_2015['away_redcards'] = away_redcards
data_2015['away_gk_saves'] = away_gk_saves
data_2015['away_pass_percentage'] = away_pass_percentage
```
