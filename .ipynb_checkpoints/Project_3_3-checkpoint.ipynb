```python
data = []

for i in training_years:
    
    number_of_weeks = 38
    
    for a in range(2, number_of_weeks + 1):
    
        fixtures_request = requests.get(f'https://v3.football.api-sports.io/fixtures?league={league_id}&season={i}&round=Regular Season - {a}', headers = headers)

        time.sleep(0.025)

        fixture_results = fixtures_request.json()['results']

        for b in range(0, fixture_results):

            home_team_id = fixtures_request.json()['response'][b]['teams']['home']['id']
            away_team_id = fixtures_request.json()['response'][b]['teams']['away']['id']

            previous_home_fixture = requests.get(f'https://v3.football.api-sports.io/fixtures?league={league_id}&team={home_team_id}&round=Regular Season - {a-1}&season={i}', headers = headers)

            time.sleep(0.025)
            
            previous_home_fixture_id = previous_home_fixture.json()['response'][0]['fixture']['id']

            if previous_home_fixture.json()['response'][0]['teams']['home']['id'] == home_team_id:

                previous_home_location = 'home'

            else:

                previous_home_location = 'away'

            previous_away_fixture = requests.get(f'https://v3.football.api-sports.io/fixtures?league={league_id}&team={away_team_id}&round=Regular Season - {a-1}&season={i}', headers = headers)

            time.sleep(0.025)
            
            previous_away_fixture_id = previous_away_fixture.json()['response'][0]['fixture']['id']

            if previous_away_fixture.json()['response'][0]['teams']['home']['id'] == away_team_id:

                previous_away_location = 'home'

            else:

                previous_away_location = 'away'

            previous_home_stats = requests.get(f'https://v3.football.api-sports.io/fixtures/statistics?fixture={previous_home_fixture_id}&team={home_team_id}', headers= headers)

            time.sleep(0.025)

            previous_away_stats = requests.get(f'https://v3.football.api-sports.io/fixtures/statistics?fixture={previous_away_fixture_id}&team={away_team_id}', headers= headers)

            if previous_home_stats.json()['response'] == [] or previous_away_stats.json()['response'] == []:

                data.append({


                    'date' : fixtures_request.json()['response'][b]['fixture']['date'],
                    'home_result' : json.dumps(fixtures_request.json()['response'][b]['teams']['home']['winner']).replace('null', 'None'),
                    'previous_home_fixture' : previous_home_fixture_id,
                    'home_team': fixtures_request.json()['response'][b]['teams']['home']['name'],
                    'home_team_id': home_team_id,
                    'away_team': fixtures_request.json()['response'][b]['teams']['away']['name'],
                    'away_team_id': away_team_id,
                    'previous_away_fixture': previous_away_fixture_id,
                    'last_home_goals': previous_home_fixture.json()['response'][0]['goals'][previous_home_location],
                    'last_away_goals': previous_away_fixture.json()['response'][0]['goals'][previous_away_location],
                    'last_home_shots_on_goal' : np.nan,
                    'last_home_shots_off_goal' : np.nan,
                    'last_home_total_shots' : np.nan,
                    'last_home_blocked_shots': np.nan,
                    'last_home_shots_insidebox': np.nan,
                    'last_home_shots_outsidebox': np.nan,
                    'last_home_fouls': np.nan,
                    'last_home_cornerkicks': np.nan,
                    'last_home_offsides': np.nan,
                    'last_home_ballpossession': np.nan,
                    'last_home_yellowcards': np.nan,
                    'last_home_redcards': np.nan,
                    'last_home_gk_saves': np.nan,
                    'last_home_pass_accuracy': np.nan,
                    'last_away_shots_on_goal' : np.nan,
                    'last_away_shots_off_goal' : np.nan,
                    'last_away_total_shots' : np.nan,
                    'last_away_blocked_shots': np.nan,
                    'last_away_shots_insidebox': np.nan,
                    'last_away_shots_outsidebox': np.nan,
                    'last_away_fouls': np.nan,
                    'last_away_cornerkicks': np.nan,
                    'last_away_offsides': np.nan,
                    'last_away_ballpossession': np.nan,
                    'last_away_yellowcards': np.nan,
                    'last_away_redcards': np.nan,
                    'last_away_gk_saves': np.nan,
                    'last_away_pass_accuracy': np.nan,


                })

            else:

                data.append({


                    'date' : fixtures_request.json()['response'][b]['fixture']['date'],
                    'home_result' : json.dumps(fixtures_request.json()['response'][b]['teams']['home']['winner']).replace('null', 'None'),
                    'previous_home_fixture' : previous_home_fixture_id,
                    'home_team': fixtures_request.json()['response'][b]['teams']['home']['name'],
                    'home_team_id': home_team_id,
                    'away_team': fixtures_request.json()['response'][b]['teams']['away']['name'],
                    'away_team_id': away_team_id,
                    'previous_away_fixture': previous_away_fixture_id,
                    'last_home_goals': previous_home_fixture.json()['response'][0]['goals'][previous_home_location],
                    'last_away_goals': previous_away_fixture.json()['response'][0]['goals'][previous_away_location],
                    'last_home_shots_on_goal' : float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][0]['value']).replace('null', '0')),
                    'last_home_shots_off_goal' : float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][1]['value']).replace('null', '0')),
                    'last_home_total_shots' : float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][2]['value']).replace('null', '0')),
                    'last_home_blocked_shots': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][3]['value']).replace('null', '0')),
                    'last_home_shots_insidebox': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][4]['value']).replace('null', '0')),
                    'last_home_shots_outsidebox': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][5]['value']).replace('null', '0')),
                    'last_home_fouls': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][6]['value']).replace('null', '0')),
                    'last_home_cornerkicks': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][7]['value']).replace('null', '0')),
                    'last_home_offsides': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][8]['value']).replace('null', '0')),
                    'last_home_ballpossession': float(previous_home_stats.json()['response'][0]['statistics'][9]['value'].strip('%')) * 0.01,
                    'last_home_yellowcards': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][10]['value']).replace('null', '0')),
                    'last_home_redcards': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][11]['value']).replace('null', '0')),
                    'last_home_gk_saves': float(json.dumps(previous_home_stats.json()['response'][0]['statistics'][12]['value']).replace('null', '0')),
                    'last_home_pass_accuracy': previous_home_stats.json()['response'][0]['statistics'][14]['value'] / previous_home_stats.json()['response'][0]['statistics'][13]['value'],
                    'last_away_shots_on_goal' : float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][0]['value']).replace('null', '0')),
                    'last_away_shots_off_goal' : float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][1]['value']).replace('null', '0')),
                    'last_away_total_shots' : float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][2]['value']).replace('null', '0')),
                    'last_away_blocked_shots': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][3]['value']).replace('null', '0')),
                    'last_away_shots_insidebox': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][4]['value']).replace('null', '0')),
                    'last_away_shots_outsidebox': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][5]['value']).replace('null', '0')),
                    'last_away_fouls': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][6]['value']).replace('null', '0')),
                    'last_away_cornerkicks': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][7]['value']).replace('null', '0')),
                    'last_away_offsides': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][8]['value']).replace('null', '0')),
                    'last_away_ballpossession': float(previous_away_stats.json()['response'][0]['statistics'][9]['value'].strip('%')) * 0.01,
                    'last_away_yellowcards': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][10]['value']).replace('null', '0')),
                    'last_away_redcards': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][11]['value']).replace('null', '0')),
                    'last_away_gk_saves': float(json.dumps(previous_away_stats.json()['response'][0]['statistics'][12]['value']).replace('null', '0')),
                    'last_away_pass_accuracy': previous_away_stats.json()['response'][0]['statistics'][14]['value'] / previous_away_stats.json()['response'][0]['statistics'][13]['value'],

                })

train_df = pd.DataFrame(data)
train_df.set_index('date', inplace=True)
train_df.loc[train_df['home_result'] == 'true', 'result'] = 1 
train_df.loc[train_df['home_result'] == 'false', 'result'] = -1
train_df.loc[train_df['home_result'] == 'None', 'result'] = 0
train_df.drop_duplicates(inplace=True)
train_df.dropna(inplace=True)
train_df.drop(columns=['home_result'], inplace=True)
train_df.sort_index(inplace=True)                                                   
```
